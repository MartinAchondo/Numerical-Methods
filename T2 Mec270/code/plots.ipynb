```python
import numpy as np
import matplotlib.pyplot as plt

x = ["Newton","Broyden","GradConj"]
e = [1e-11, 7e-10,1e-6]
e2 = [3e-15, 3e-8,1e-7]

fig , ax1 = plt.subplots()

ax1.plot(x, e, c='b', ls='-', marker='s', label='Error')
ax1.set_yscale('log')
ax1.grid()
ax1.set_xlabel('Métodos')
ax1.set_ylabel('Error')


ax2 = ax1.twinx()

ax2.plot(x, e2, c='r', ls='-', marker='o', label='Residual')
ax2.set_ylabel('Residual')
ax2.set_yscale('log')

ax2.legend(loc='upper left')
ax1.legend(loc = 'lower right')

```
