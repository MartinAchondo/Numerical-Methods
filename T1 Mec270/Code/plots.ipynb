```python
import numpy as np
import matplotlib.pyplot as plt

x = ["Chol","LU","QR","Jac","Gauss","Grad"]
e = [1e-10, 1e-11,1e-10,1e-5, 1e-6, 1e-8]
e2 = [6.5e-4, 10e-4,6e-4,21e-4, 19e-4, 2.9e-4]
e3 = [2000000,7000000, 1900,  196000, 196000, 94000]

fig , ax1 = plt.subplots()

ax1.plot(x, e, c='r', ls='-', marker='s', label='Residuales')
ax1.set_yscale('log')
ax1.grid()
ax1.set_xlabel('Métodos')
ax1.set_ylabel('Error')


ax2 = ax1.twinx()

ax2.plot(x, e3, c='b', ls='-', marker='o', label='Flops')
ax2.set_ylabel('flops')
ax2.set_yscale('log')

ax1.legend(loc='upper left')
ax2.legend(loc = 'lower right')
```
